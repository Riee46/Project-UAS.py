**Function** memiliki argumen atau masukkan pada parameter sehingga hasil yang didapatkan dapat dinamis sesuai dengan masukkan.

```python
#FUNCTION TANPA ARGUMEN#

def perkenalan():
  print("halo aku sapta")

perkenalan()



```

```python
#FUNCTION DENGAN ARGUMEN PT.2#

def asprak_algo(asprak):
  for i in asprak:
    print(i)

list_asprak = ["Sapta","Ariski","Tsaqif","Irvan","Fauzul","Athar","Eka","Rashtian","Fathan","Taufiq","Vian","Zaka"]

print(asprak_algo(list_asprak))
```

### **DEFAULT PARAMETER**

### **Return/Kembali**

Suatu function dapat memberikan nilai kembalian(Return) maupun tidak ketika function tersebut dipanggil.

Tujuan dari suatu function memberikan nilai kembalian(Return) agar hasil pemrosesan dapat diolah kembali pada prosedur selanjutnya.

```python
def dengan_return(a, b):
    return a + b

# Versi 1
dengan_return(10,11)

# Versi 2
print(dengan_return(10,11))
```

```python
def tidak_ada_return(a, b):
    result = a + b

# Versi 1
tidak_ada_return(10,11)

# Versi 2
print(tidak_ada_return(10,11))
```

###Variabel Global dan Lokal

Variabel global adalah variabel yang dapat diakses dari mana saja

Variabel Lokal adalah variabel yang hanya dapat diakses dari sebuah function saja

```python
Variabel_Global = "ini variabel global"

def sebuahfungsi():
  Variabel_lokal = "Ini variabel lokal"

print(Variabel_Global)
print(Variabel_lokal) #Variabel lokal tidak dapat diakses karena berada di dalam function
```

```python
Variable_Global = "ini variabel global"

def sebuahfungsi():
  global Variable_Global #Mengubah variabel lokal menjadi global
  Variable_Global = "Ini variabel lokal"

print(Variable_Global)

sebuahfungsi() #harus memanggil fungsi terlebih dahulu

print(Variable_Global) #Variabel global diupdate sesuai dengan variabel lokal
```

### **LAMBDA**


**Lambda** adalah fungsi anonim tanpa nama yang digunakan untuk membuat fungsi sederhana dalam konteks tertentu. Mereka lebih ringkas dan hanya memiliki satu ekspresi.

Beberapa bahasa pemrograman dengan fungsi Lambda:
- Java
- JavaScript
- Python
- PHP

```python
x = lambda a : a + 10
print(x(5))
```

```python
x = lambda a, b : a * b
print(x(5, 6))
```

```python
x = lambda a, b, brian : a + b + brian
print(x(5, 6, 2))
```

```python
x = lambda nama_depan,nama_belakang: f'nama lengkapku {nama_depan} {nama_belakang}'
print(x('Ariski','Raharjo'))
```

### **Eror Handling**
**Eror handling** adalah suatu fitur pada python untuk menangani sebuah eror pada program


*   **try** Digunakan untuk mencoba suatu blok kode apakah terjadi sebuah eror
*   **except** digunakan untuk menangani apabila terjadi error.

```python
sebuah_variable = "Bukan Int"

try:
  sebuah_variable = int(sebuah_variable)
except ValueError:
  print("Variable yang di casting harus berupa angka")
```

```python
def Penjumlahan(Angka1,Angka2):
  try:
    Angka1 = int(Angka1)
    Angka2 = int(Angka2)
    hasil = Angka1 + Angka2
    return hasil
  except ValueError:
    return "Input harus berupa angka"

angka1 = input("Masukan angka 1 : ")
angka2 = input("Masukan angka 2 : ")

Penjumlahan(angka1,angka2)
```

## SOAL 1
Buatkan Sebuah Function yang memiliki fungsi untuk mencari nilai rata - rata tanpa menggunakan fungsi Sum() dengan inputan parameter dalam bentuk list

```python
def hitung_rata_rata(data):
    if not data:
        return 0

    total = 0
    count = 0

    for angka in data:
        total += angka
        count += 1

    rata_rata = total / count
    return rata_rata

data_list = [10, 20, 30, 40 ,50]
rata_rata = hitung_rata_rata(data_list)
print("Nilai rata-rata:",rata_rata)
 
```

## SOAL 2
Buatkan Sebuah Function untuk mengembalikan nilai terbesar dari sebuat list tanpa menggunakan Function Sum

```python
def cari_nilai_terbesar(data):
    if not data:  
    
        return None  
    terbesar = data[0]  
    
    for angka in data:
        if angka > terbesar:  
            terbesar = angka  
    
    return terbesar

data_list = [10, 20, 5, 40, 30]
nilai_terbesar = cari_nilai_terbesar(data_list)
print("Nilai terbesar:", nilai_terbesar)
```

## SOAL 3
Buatlah Sebuah Function untuk menghitung Faktorial yang menerima satu parameter

```python
def hitung_faktorial(n):
    if n < 0:
        return "Faktorial tidak terdefinisi untuk bilangan negatif"
    elif n == 0 or n == 1:
        return 1  
    else:
        faktorial = 1
        for i in range(2, n + 1):
            faktorial *= i  
        return faktorial

angka = 5
hasil_faktorial = hitung_faktorial(angka)
print(f"Faktorial dari {angka} adalah: {hasil_faktorial}")
```
