# OPERASI DATA LIST

```python
# append
Nama_Buah = ["Apel", "Mangga", "Nanas", "Manggis"]
Nama_Buah.append('Kiwi')
print(Nama_Buah)
```

```python
# insert
Nama_Buah = ["Apel", "Mangga", "Nanas", "Manggis"]
Nama_Buah.insert(1, 'Rambutan') #memasukkan nilai dengan mengakses indexnya
print(Nama_Buah)
```

```python
# remove
Nama_Buah = ["Apel", "Mangga", "Nanas", "Manggis"]
Nama_Buah.remove('Apel')
print(Nama_Buah)
```

```python
# pop
Nama_Buah = ["Apel", "Mangga", "Nanas", "Manggis"]
Nama_Buah.pop(1) # menghapus value dengan index tertentu
print(Nama_Buah)
```

```python
# extend
Nama_Buah = ["Apel", "Mangga", "Nanas", "Manggis"]
List_Baru = ["Anggur", "Sawo", "Semangka"]
List1 = ["Ayam","Sayur"]
# Nama_Buah = Nama_Buah + List_Baru + List1
Nama_Buah.extend(List_Baru)
# Nama_Buah.extend()
print(Nama_Buah)
```

```python
# del (menghapus value dengan index tertentu)
Nama_Buah = ["Apel", "Mangga", "Nanas", "Manggis"]
del Nama_Buah[2:4]
print(Nama_Buah)
```

```python
# del (tanpa index)
Nama_Buah = ["Apel", "Mangga", "Nanas", "Manggis"]
del Nama_Buah
print(Nama_Buah)
```

```python
# copy
Nama_Buah = ["Apel", "Mangga", "Nanas", "Manggis"]
List_Terbaru = Nama_Buah.copy()
print(List_Terbaru)
```

```python
# reverse
Nama_Buah = ["Apel", "Mangga", "Nanas", "Manggis"]
Nama_Buah.reverse()
print(Nama_Buah)
```

```python
# clear
Nama_Buah = ["Apel", "Mangga", "Nanas", "Manggis"]
Nama_Buah.clear()
print(Nama_Buah)
```

```python
# sort
Nama_Buah = ["Apel", "Mangga", "Nanas", "Manggis", "Ceri", "Delima"]
Nama_Buah.sort()
print(Nama_Buah)
```

# Dictionary

```python
dictionary = {
    "Prodi" : "Teknologi Informasi",
    "Prodi" : "FKIP",
    "Fakultas" : "Ilmu Komputer",
    "Univ" : "Universitas Jember",
    "Tahun Masuk" : 2024,
    "Tahun Sekarang" : 2024,
}
# dictionary["Prodi"] = "kedokteran hewan"
# print(f'{dictionary["Prodi"]}')
print(dictionary)
```

**Disini terdapat perbedaan antara List, Set, Tuple dan Dictionary**
- List: Dapat di ubah, Dapat dipanggil dengan urutan indeks, mengizinkan duplikat.
- Set: Tidak dapat di ubah, Tidak dapat dipanggil dengan urutan indeks, tidak mengizinkan duplikat.
- Tuple: Tidak dapat di ubah, Dapat dipanggil dengan urutan indeks, mengizinkan duplikat.
- Dictionary: Dapat diubah, dipanggil menggunakan kunci (key) bukan indeks, tidak mengizinkan duplikat kunci(key) tetapi mengizinkan duplikat nilai(value).

# Operator 

Di python bisa ngitung-ngitung ga bang? Bisa banget le, operasi dasar bilangan seperti penjumlahan, pengurangan, perkalian, dan operasi lainnya bisa dilakukan dalam python. Bahkan operasi selain itung-itungan juga bisa. Contohnya seperti kode di bawah ini

### Operator Aritmatika

```python
penjumlahan = 10 + 5
print(f'Hasil penjumlahan: {penjumlahan}')

pengurangan = 10 - 5
print(f'Hasil pengurangan: {pengurangan}')

perkalian = 10 * 5
print(f'Hasil perkalian: {perkalian}')

pembagian = 5 / 2
print(f'Hasil pembagian: {pembagian}')

pembagianBulat = 5 // 2
print(f'Hasil pembagian bulat: {pembagianBulat}')

pembagianBulat = 5 // 2
print(f'Hasil pembagian bulat: {pembagianBulat}')

perpangkatan = 10 ** 2 #Pangkat 2
print(f'Hasil perpangkatan: {perpangkatan}')

modulus = 11 % 2
print(f'Hasil modulus: {modulus}')
```

### Prioritas Operasi

- Prioritas pengerjaan operasi
1. () / tanda kurung
2. eksponen
3. perkalian, pembagian, modulus, floor division
4. penjumlahan, pengurangan

```python
a = 3
b = 4
c = 5

hasil = a ** b * c + a // c % b / a
print(f"a ** b * c + a // c % b / a =", hasil)

#Tanpa tanda kurung
hasil = a + b * c
print("a + b * c =", hasil)

#Menggunakan tanda kurung
hasil = (a + b) * c
print("(a + b) * c =", hasil)
```

### Operator Keanggotaan

```python
orangDingin = ["Sapta", "Ariski", "Vian", "Irvan"]

# Bernilai true jika sebuah data terdapat dalam suatu list
print("Sapta" in orangDingin) 

# Bernilai false jika sebuah data tidak terdapat dalam suatu list
print("Taufik" in orangDingin)

# Bernilai true jika sebuah data tidak berada dalam suatu list
print("Athar" not in orangDingin)

# Bernilai false jika sebuah data berada dalam list
print("Vian" not in orangDingin)
```

### Operator Logika

```python
# NOT (Jika true menjadi false dan sebaliknya)
a = True
b = not a 
print("Data a:", a)
print("Data b:", b)

# OR (Minimal salah satu nilai True, maka hasilnya True )
a = True
b = False
c = a or b
print(a, "OR", b, "=", c)

# AND (Jika dua buah nilai True, maka hasilnya True)
a = True
b = False
c = a and b
print(a, "AND", b, "=", c)

# XOR (Akan True jika dua buah nilai nya berbeda alias tidak boleh sama)
a = True
b = True
c = a ^ b
print(a, "XOR", b, "=", c)
```

```python
first_name = "Tatang"
last_name = "Iskandar"

#1 operator penjumlahan
namaku = first_name + " " + last_name
print(namaku)

#2 operator penugasan
first_name += last_name
print(first_name)

#3 menjadikan sebuah list dan digabung
namaku = " ".join([first_name,last_name])
print(namaku)

#4 format string (Python model)
namaku = "{} {}".format(first_name,last_name)
print(namaku)

#5 format string (f-strings)
namaku = f"{first_name} {last_name}"
print(namaku)

#6 format string (C model)
namaku = "%s %s" % (first_name,last_name)
print(namaku)

#6 menggabungkan string literal
print("Fathan" + "Sangat" + "Dingin")
print("Fathan" "Dingin")
```
